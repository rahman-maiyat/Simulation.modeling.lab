```python

import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Create My Personal Matrix

```python
d1 = 2   #last digit of my student ID = 2
d2 = 2    #second last digit of my student ID = 2

A = np.array([[d1+2, d2+1], [2*d1, d2+2]])
A = np.array([[4, 3], [4, 4]])
```

2. Analyze Matrix A

Using NumPy, compute:

● Shape of A

● Determinant

● Rank

● Eigenvalues

● Inverse (only if determinant ≠ 0)

```python
A = np.array([[4, 3],[4, 4]])


size_A = A.shape
det_A = np.linalg.det(A)
rank_A = np.linalg.matrix_rank(A)
eigenvalues = np.linalg.eig(A)


if det_A != 0:
    inv_A = np.linalg.inv(A)
else:
    inv_A = "Matrix is singular, inverse does not exist"


print("Matrix Size:", size_A)
print("Determinant:", det_A)
print("Rank:", rank_A)
print("Eigenvalues and Eigenvectors:\n", eigenvalues)
print("Inverse:\n", inv_A)
```
